```python
import itertools
import pandas as pd
from collections import Counter
```

```python
train = pd.read_json('../data/train.json', encoding='utf-8')
train.shape
```

```python
validation = pd.read_json('../data/val.json', encoding='utf-8')
validation.shape
```

```python
song_meta = pd.read_json('../data/song_meta.json', encoding='utf-8')
song_meta.head()
```

```python
song_meta['year'] = song_meta.issue_date.astype('str').apply(lambda x: x[:4]).astype('int')
song_meta.head()
```

```python
year_index = sorted(pd.unique(song_meta.year).tolist())
year_index
```

```python
year_to_index = dict()
index_to_year = dict()

for i, year in enumerate(year_index):
    year_to_index[year] = i
    index_to_year[i] = year
    
print(f"year to index: {year_to_index}")
print()
print(f"index to year: {index_to_year}")
```

```python
song_year = dict(zip(song_meta['id'], song_meta['year']))
song_year
```

```python
train['year'] = train.songs.apply(lambda e: Counter(list(map(lambda x: song_year[x], e))))
train.head()
```

```python
validation['year'] = validation.songs.apply(lambda e: Counter(list(map(lambda x: song_year[x], e))))
validation.head()
```

```python
import numpy as np

year_onehot = np.zeros((train.shape[0] + validation.shape[0], len(year_to_index)))
year_onehot.shape
```

```python
merged_df = pd.concat([train, validation], axis=0)
merged_df.shape
```

```python
from tqdm.auto import tqdm

for idx, row in tqdm(enumerate(merged_df.iterrows()), total=merged_df.shape[0]):
    try:
        hot = sorted([ year_to_index[year] for year in list(row[-1].year.keys()) ])
        year_onehot[idx, hot] = 1
    except:
        print(row)
    
year_onehot[:10]
```

```python
np.save('../data/year_onehot.npy', year_onehot)
```
